## Create Schemaless Delta Table to Ingest

```python
%sql
create table if not exists dev.demo_db.invoices_raw_1;
```

```python
%sql describe extended dev.demo_db.invoices_raw_1
```

```python
%sql
COPY INTO dev.demo_db.invoices_raw_1
FROM
(select * from "/Volumes/dev/demo_db/landing")
FILEFORMAT = CSV
FORMAT_OPTIONS ('header' = 'true',"inferSchema" = "true","mergeSchema" = "true","timestampFormat" = "d-M-y H.m")
COPY_OPTIONS ("mergeSchema" = "true")
```
